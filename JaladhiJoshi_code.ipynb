```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from dateutil.relativedelta import relativedelta
df = pd.read_csv('data.csv')
#this is question1 loanterm is converted in months and then added to the original date loanmonths is the list that has the loan term in months form
df['originationDate'] = pd.to_datetime(df['originationDate'], format='%m/%Y')
df['loanmonths']= df['loanTerm'].str.extract('(\d+)').astype(float)
for i in range(len(df['loanTerm'])):
    if pd.isna(df['loanTerm'][i]):
          df.loc[i,'loanmonths']=0
    elif 'y' in df['loanTerm'][i]:
          df.loc[i,'loanmonths']=df['loanmonths'][i]*12
maturity_dates = []
for i in range(len(df)):
     maturity_dates.append(df['originationDate'][i] + relativedelta(months=df['loanmonths'][i]))

df['maturityDate'] = maturity_dates
print(df['maturityDate'])
d1 = datetime(2022,12,1)  
#question2 the current month and year is dec22 so remaining term is calculated by the differencein months between current date and maturity date
for i in range(len(df['loanTerm'])):
    
    print((df['maturityDate'][i].year-d1.year)*12+df['maturityDate'][i].month-d1.month)

#question3 partA states is a dictionary whose items has two values,first one keeping the loan count and second loan amount which is sorted and then 2nd element is the required element
states={}
for i in range(len(df)):
    if df['obligorState'][i] not in states:
        states[df['obligorState'][i]]=[1,df['loanAmount'][i]]
    else:
        states[df['obligorState'][i]][0]+=1
        states[df['obligorState'][i]][1]+=df['loanAmount'][i]
done = sorted(states.items(), key=lambda x:(-x[1][0]))
print(done[1][0])
# #question3 partB 
dum=sorted(states.items(), key=lambda x:(-x[1][1]))
print(dum[1][0])
#question4 percenta dictionary's items has two values,first one keeping the used vehicle loans and second one keeping the new vehicle loans and percentage is used/used+new
percenta={}
for i in range(len(df)):
    if df['originationDate'][i].month not in percenta:
        if df['vehicleNewUsedCode'][i]=="New":
            percenta[df['originationDate'][i].month]=[0,1]
        else:
            percenta[df['originationDate'][i].month]=[1,0]
    else:
        if df['vehicleNewUsedCode'][i]=="New":
            percenta[df['originationDate'][i].month][1]+=1       
        else:
            percenta[df['originationDate'][i].month][0]+=1
percentag={}
for i in range(1,13):
    percentag[i]=((percenta[i][0]/((percenta[i][1])+(percenta[i][0]))))*100
print(percentag)
plt.plot(percentag.keys(),percentag.values())
plt.xticks(np.arange(1,13,1))
plt.show()
#question5 if beginning loan balance and ending loan balance are same, the loan is defaulted divided by the total loans give the percent of defaulted loans
df['defaultFlag'] = False
count=int(0)
for i in range(len(df)):
    if (df['beginningLoanBal'][i]==df['endingLoanBal'][i]):
        df.loc[i,'defaultFlag']=True
        count+=1
  
print(df['defaultFlag'])
print(count*100/50043)
#question 6 interest rate is scheduled interest/scheduled principal amout
counter=int(0)
req=int(0)
for i in range(len(df)):
    if(df['vehicleNewUsedCode'][i]=="Used") and (df['coObligorIndicator'][i]==True):
        
        counter=counter+1
        if (df['scheduledPrincipalAmount'][i] != 0) and (df['scheduledInterestAmount'][i]/df['scheduledPrincipalAmount'][i] <= 0.03):
            req=req+1
print(req*100/counter) 
#question 7 model dictionary keys are the vehicle model name and vehicle model year and items are their count in FL
model={}
for i in range(len(df)):
    if(df['obligorState'][i]=="FL"):
        if (df['vehicleModelName'][i],df['vehicleModelYear'][i]) not in model:
            
                 model[df['vehicleModelName'][i],df['vehicleModelYear'][i]]=1
           
        else:
            
                 model[df['vehicleModelName'][i],df['vehicleModelYear'][i]]+=1    

sort_model=sorted(model.items(),key=lambda x:(-x[1]))
print(sort_model[0])
#question 8 obligorcreditscore is divided in 5 parts using cut and each groups intrate is averaged
df['credit_score_buckets'] = pd.cut(df['obligorCreditScore'], bins=5)
bucket_df = df.groupby('credit_score_buckets')['intRate'].mean()

# Plot the results using a bar graph
bucket_df.plot(kind='bar')
plt.show()
```
